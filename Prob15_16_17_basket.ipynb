```python
import pandas as pd
from apyori import apriori

# Load the dataset
file_path = "D:\downloaads\ML Dataset\LP-1 Dataset\Apriori\Order1.csv"
data = pd.read_csv(file_path)

# Step 1: Data Pre-processing
# Convert the data into a list of lists for each transaction
transactions = data.groupby(['Member_number', 'Date'])['itemDescription'].apply(list).tolist()

# Step 2: Train Apriori on the dataset
# Using apriori with minimum support and confidence
rules = apriori(transactions, min_support=0.01, min_confidence=0.02, min_lift=1, min_length=2)

# Step 3: Visualize the results
# Convert rules to a list for easy visualization
results = list(rules)

# Print the results
print(len(results))
print(results)
for rule in results:
    items = [x for x in rule.items]
    print(f"Rule: {items}")
    print(f"Support: {rule.support}")
    for ordered_stat in rule.ordered_statistics:
        print(f"Confidence: {ordered_stat.confidence}")
        print(f"Lift: {ordered_stat.lift}")
    print("="*20)
```

```python
import pandas as pd
from apyori import apriori

# Load the dataset
file_path = "D:\downloaads\ML Dataset\LP-1 Dataset\Apriori\Oder3.csv"
data = pd.read_csv(file_path)

# Step 1: Data Pre-processing
# Fill NaN values with the most frequent item in each column
data = data.applymap(str)
data = data.apply(lambda x: x.fillna(x.value_counts().idxmax()), axis=0)

# Step 2: Generate the list of transactions
transactions = data.apply(lambda row: [item for item in row if str(item) != 'nan'], axis=1).tolist()

# Step 3: Train Apriori Algorithm
# Set parameters for support, confidence, and lift
min_support = 0.001  # Adjust based on dataset size
min_confidence = 0.2
min_lift = 2
min_length = 2

# Run the apriori algorithm
rules = apriori(transactions, min_support=min_support, min_confidence=min_confidence, min_lift=min_lift, min_length=min_length)

# Step 4: Visualize the list of rules
# Convert rules to a list for easy visualization
results = list(rules)
print(len(results))

# Print the results
for rule in results:
    items = [x for x in rule.items]
    print(f"Rule: {items}")
    print(f"Support: {rule.support}")
    for ordered_stat in rule.ordered_statistics:
        print(f"Confidence: {ordered_stat.confidence}")
        print(f"Lift: {ordered_stat.lift}")
    print("="*20)
```
